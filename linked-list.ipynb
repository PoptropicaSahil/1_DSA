```python
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
    
class Solution:

    def __init__(self, head: Node) -> None:
        self.head = head

    def traverse_list(self) -> None:
        print('traversing ...')
        start = self.head
        while start is not None:
            print(f"Now at {start.data}, next is {start.next}")
            start = start.next
        print('\n')


    def add_new_head(self, new_head: Node) -> None:
        new_head.next = self.head
        self.head = new_head

    def search_node(self, given_node: Node) -> bool:
        start = self.head
        while start is not None:
            if start.data == given_node.data:
                return True
            start = start.next
        return False
    
    def delete_node(self, given_node: Node) -> None:
        start = self.head

        while start is not None:

            if start.next == given_node: # if found the node
                if start.next.next is not None:
                    start.next = start.next.next
                else:
                    start.next = None
            start = start.next
    
    def delete_tail(self) -> None:
        start = self.head
        while start.next.next is not None:
            start = start.next
        start.next = None

    def reverse_list(self) -> None:
        # initialise
        prev = None
        current = self.head
        while current is not None:
            # We have to do current.next = prev
            # and shift prev, current forward
            # Just that befire doing current.next = prev, we have to save the next
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    def update_node(self, old_node: Node, new_node: Node) -> None:
        start = self.head
        while start is not None:
            if start.data == old_node.data:
                start.data = new_node.data
            start = start.next



family_father = Node('Bob')
family_mother = Node('Jane')
family_child1 = Node('John')
family_child2 = Node('Mary')

family = Solution(head = family_father)

family_father.next = family_mother
family_mother.next = family_child1
family_child1.next = family_child2

print('traverse list')
family.traverse_list()


family_grandfather = Node('Kaka')
family.add_new_head(new_head = family_grandfather)
print(f"added new head = {family_grandfather}")
family.traverse_list()


family.search_node(given_node = family_child2)
family.search_node(given_node = Node('popo'))

print(f"deleting node father ")
family.delete_node(given_node = family_father)
family.traverse_list()

print(f"deleting tail ")
family.delete_tail()
family.traverse_list()


print(f"reversing list ")
family.reverse_list()
family.traverse_list()

print(f"reversing list ")
family.reverse_list()
family.traverse_list()
```

```python
class Node:
    def __init__(self, data, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev


class Solution:
    def __init__(self, head: Node) -> None:
        self.head = head

    def traverse_list(self) -> None:
        print("Traversing forward ...")
        start = self.head
        while start is not None:
            print(
                f"Now at {start.data}, next is {start.next.data if start.next else None}"
            )
            start = start.next
        print("\n")

    def traverse_list_reverse(self) -> None:
        print("Traversing backward ...")
        start = self.head
        # Go to the tail of the list
        while start.next is not None:
            start = start.next

        # Now traverse backward
        while start is not None:
            print(
                f"Now at {start.data}, prev is {start.prev.data if start.prev else None}"
            )
            start = start.prev
        print("\n")

    def add_new_head(self, new_head: Node) -> None:
        if self.head is not None:
            self.head.prev = new_head
        new_head.next = self.head
        self.head = new_head

    def search_node(self, given_node: Node) -> bool:
        start = self.head
        while start is not None:
            if start.data == given_node.data:
                return True
            start = start.next
        return False

    def delete_node(self, given_node: Node) -> None:
        start = self.head

        while start is not None:
            if start.data == given_node.data:
                if start.prev is not None:
                    start.prev.next = start.next
                else:
                    self.head = start.next  # If we're deleting the head
                if start.next is not None:
                    start.next.prev = start.prev
                return  # Exit once the node is found and deleted
            start = start.next

    def delete_tail(self) -> None:
        start = self.head
        if start is None:  # If the list is empty
            return
        if start.next is None:  # If the list has only one node
            self.head = None
            return

        # Traverse to the last node
        while start.next is not None:
            start = start.next

        # Remove the last node
        start.prev.next = None

    def reverse_list(self) -> None:
        current = self.head
        prev_node = None
        while current is not None:
            # Swap the next and prev pointers
            next_node = current.next
            current.next = prev_node
            current.prev = next_node

            # Move prev_node and current one step forward
            prev_node = current
            current = next_node

        # Update the head of the list to be the last non-None node encountered
        if prev_node is not None:
            self.head = prev_node

    def update_node(self, old_node: Node, new_node: Node) -> None:
        start = self.head
        while start is not None:
            if start.data == old_node.data:
                start.data = new_node.data
            start = start.next


# Example Usage:
family_father = Node("Bob")
family_mother = Node("Jane")
family_child1 = Node("John")
family_child2 = Node("Mary")

family = Solution(head=family_father)

family_father.next = family_mother
family_mother.prev = family_father
family_mother.next = family_child1
family_child1.prev = family_mother
family_child1.next = family_child2
family_child2.prev = family_child1

family.traverse_list()

# Add a new head
print('adding new head robert')
family_grandfather = Node("Robert")
family.add_new_head(family_grandfather)
family.traverse_list()

# Delete a node
print("deleting mother node")
family.delete_node(family_mother)
family.traverse_list()

# Delete the tail
print("deleting tail")
family.delete_tail()
family.traverse_list()

# Reverse the list
print("reversing list")
family.reverse_list()
family.traverse_list()

# Update a node
print("updating node")
family.update_node(family_father, Node("Robert Sr."))
family.traverse_list()
```
